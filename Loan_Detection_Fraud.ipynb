For the nominal data I am going with get_dummies.

```python
x=pd.get_dummies(x,columns=['Gender','Race','Employment_Type','Citizenship_Status','Language_Proficiency',
                              'Disability_Status','Criminal_Record','Zip_Code_Group'],drop_first=True)
```

For the ordinal data I will go with Ordinal Encoding.  
>I am not using label encoder as it will give order alphabetically like  
Bachelor's = 0  
Graduate   = 1  
High School = 2  
Some College = 3  
which is wrong, so In **ordinal encoder** we will provide the order.


```python
from sklearn.preprocessing import OrdinalEncoder

education_order = [['High School', 'Some College', "Bachelor's", 'Graduate']]
ordinal=OrdinalEncoder(categories=education_order)
x[['Education_Level']]=ordinal.fit_transform(x[['Education_Level']])
```

```python
x['Education_Level']
```

Our target variable is in form of 'Approved, Declined' so we convert it into 1,0.

```python
type(y)
```

```python
y=y.map({'Approved':1,'Denied':0})
```

Before moving ahead drop the age_group column as it is redundant and the original age column will give us more information for our model fitting.

```python
from sklearn.model_selection import train_test_split,StratifiedKFold,cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix 
```

StratifiedKFold is a variation of K-fold which returns stratified folds: each set contains approximately the same percentage of samples of each target class as the complete set.  
stratify=y: Keeps the proportion of Approved and Denied balanced in both sets (important for imbalanced classification)

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE(random_state=42)
x_resampled,y_resampled=sm.fit_resample(x_train,y_train)
```

```python
scaler=StandardScaler()
x_train_norm=scaler.fit_transform(x_resampled)
x_test_norm=scaler.transform(x_test)
```

`class_weight='balanced'` tells the model:

â€œClass 0 (Denied) and Class 1 (Approved) are imbalanced â€” give extra weight to the minority class.â€

```python
log_model=LogisticRegression(class_weight='balanced')
log_model.fit(x_train_norm,y_resampled)
log_pred=log_model.predict(x_test_norm)
log_accuracy=accuracy_score(y_test,log_pred)
log_classification_report=classification_report(y_test,log_pred,target_names=['Denied','Approved'])
print(f'Accuracy score is {log_accuracy}')
print(60*'-')
print(f'Classification report is \n{log_classification_report}')
```

**Overall Accuracy**:  
62.26% of total predictions were correct â€” not bad for a first baseline, but thereâ€™s room to improve.    

**Precision (Approved = 0.56)**:  
Out of all the "Approved" predictions, only 56% were actually correct.  

**Recall (Approved = 0.62)**:  
Out of all actual "Approved" loans, the model caught 62% â€” this is not bad.  

**F1-score**:  
A balance between precision and recall.Approved class has lower F1 than Denied â€” indicating that the model struggles more with predicting Approvals, which is your minority class.



```python
skf=StratifiedKFold(n_splits=5,shuffle=True,random_state=42)
log_score=cross_val_score(log_model,x_train_norm,y_resampled,cv=skf,scoring='balanced_accuracy').mean()
log_f1=cross_val_score(log_model,x_train_norm,y_resampled,cv=skf,scoring='f1').mean()
log_precision=cross_val_score(log_model,x_train_norm,y_resampled,cv=skf,scoring='precision').mean()
log_recall=cross_val_score(log_model,x_train_norm,y_resampled,cv=skf,scoring='recall').mean()
log_roc_auc=cross_val_score(log_model,x_train_norm,y_resampled,cv=skf,scoring='roc_auc').mean()

print(f'balanced accuracy {log_score}')
print(f'F1 score {log_f1}')
print(f'precision is {log_precision}')
print(f'recall is {log_recall}')
print(f'roc_auc_score is {log_roc_auc}')
```

The mean CV accuracy is ~62.13%, which is consistent with your earlier results.  

There's low variance between the folds (range â‰ˆ 1.7%), suggesting that your model is stable and not overfitting on any specific fold.

```python
models={'RandomForestClassifier':RandomForestClassifier(class_weight='balanced',random_state=42),
        'SVC':SVC(class_weight='balanced',probability=True,random_state=42),
        'KNN':KNeighborsClassifier(),
        'DecisionTree':DecisionTreeClassifier(class_weight='balanced',random_state=42),
        'Gradientboosting':GradientBoostingClassifier(random_state=42),
        'Xgboost':XGBClassifier(scale_pos_weight=1.3, use_label_encoder=False,
                             eval_metric='logloss', random_state=42),
        'Adaboost':AdaBoostClassifier(random_state=42)}
```

```python

model_list=[]
score=[]
f1s=[]
precisions=[]
recalls=[]
roc_aucs=[]

for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(x_train_norm,y_resampled)
    y_pred=model.predict(x_test_norm)

    accuracy=cross_val_score(model,x_train_norm,y_resampled,cv=skf,scoring='balanced_accuracy').mean()
    f1=cross_val_score(model,x_train_norm,y_resampled,cv=skf,scoring='f1').mean()
    precision=cross_val_score(model,x_train_norm,y_resampled,cv=skf,scoring='precision').mean()
    recall=cross_val_score(model,x_train_norm,y_resampled,cv=skf,scoring='recall').mean()
    roc_auc=cross_val_score(model,x_train_norm,y_resampled,cv=skf,scoring='roc_auc').mean()

    model_name=list(models.keys())[i]
    model_list.append(model_name)
    score.append(accuracy)
    f1s.append(f1)
    precisions.append(precision)
    recalls.append(recall)
    roc_aucs.append(roc_auc)

report=pd.DataFrame()
report['model']=model_list
report['balanced_score']=score
report['f1']=f1s
report['precision']=precisions
report['recall']=recalls
report['roc_auc_score']=roc_aucs

report










```
